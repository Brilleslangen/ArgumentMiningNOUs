```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns

import sys
sys.path.append('..')
from util.helpers import labels, id2label
```

```python
# Read the data from the CSV file
df_nou = pd.read_csv('./BERT_inference_dataset/nou_predictions_of_test.csv')
df_nou = df_nou.drop(columns=["Unnamed: 0","text"])
df_nou
```

```python
# Create the confusion matrix

y_true = df_nou["label"]
y_pred = df_nou["prediction"]

# Transform the labels to textual
y_true = [id2label(y) for y in y_true]
y_pred = [id2label(y) for y in y_pred]
label = labels()

conf_mat = confusion_matrix(y_true, y_pred)

sns.heatmap(conf_mat, annot=True, fmt='d', cmap="Blues", xticklabels=label, yticklabels=label)
```
