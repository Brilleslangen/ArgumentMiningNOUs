```python
df  = pd.read_csv('../dataset/nou_hearings.csv')
argument_df = pd.DataFrame(columns=['paragraph', 'Response'])

df
```

```python
counter = 0

prompt = "Please provide the main arguments and evidence presented in the text, without any paraphrasing or summarization, and with the original wording of the text maintained: "

for row in df.itertuples():
    chat_completion = openai.chat.completions.create(
        model="meta-llama/Llama-2-70b-chat-hf",
        messages=[{"role": "user", "content": prompt + row[2]}]
    )
    print(chat_completion.choices[0].message.content)
    argument_df.loc[len(argument_df)] = [row[2], chat_completion.choices[0].message.content]
    print(counter)

argument_df
```
