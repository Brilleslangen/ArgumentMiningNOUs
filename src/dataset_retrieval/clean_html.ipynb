```python
import pandas as pd
```

```python
df = pd.read_csv('../../dataset/html_hearings.csv')
df
```

```python
# Delete the rows only containing len
df2 = df[df['paragraph'].str.len() >= 40].reset_index(drop=True)
df2 = df2[df2["paragraph"].str.contains("http") == False].reset_index(drop=True) # Remove rows with links
df2
```

```python
# save to new csv
df2.to_csv('../../dataset/html_hearings_filtered.csv', index=False)
```
