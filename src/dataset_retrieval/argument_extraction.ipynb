```python
counter = 0
df  = pd.read_csv('../../dataset/nou_hearings.csv')
argument_df = pd.read_csv('argument_df.csv')

prompt = "I have a hearing from an organisation and want to extract the arguments in this text, write the arguments without any paraphrasing and summarization. Write the output in norwegian: "

for row in df.itertuples():
    counter += 1
    print(counter)
    if counter <= 1586:
        continue

    chat_completion = openai.chat.completions.create(
        model="meta-llama/Llama-2-70b-chat-hf",
        messages=[{"role": "user", "content": prompt + row[2]}]
    )
    print(chat_completion.choices[0].message.content)
    argument_df.loc[len(argument_df)] = [row[2], chat_completion.choices[0].message.content]
    print(counter)
    

argument_df
```
